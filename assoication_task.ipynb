```python
# Replace all values that are not 0 to binary value of 1
for i, row in df2.iterrows():
    murder = df2.at[i, 'Murder']
    violet = df2.at[i, 'Violent Total']
    robbery = df2.at[i, 'Robbery']
    assualt = df2.at[i, 'Aggravated Assault']
    property = df2.at[i, 'Property Total']
    burglary = df2.at[i, 'Burglary']
    larceny = df2.at[i, 'Larceny']
    motor = df2.at[i, 'Motor Vehicle Theft']
    if(murder != 0):
        #print("Non zero value")
        df2.at[i, 'Murder'] = 1
    if(violet != 0):
        #print("Non zero value")
        df2.at[i, 'Violent Total'] = 1
    if(robbery != 0):
        #print("Non zero value")
        df2.at[i, 'Robbery'] = 1
    if(assualt != 0):
        3print("Non zero value")
        df2.at[i, 'Aggravated Assault'] = 1
    if(property != 0):
        #print("Non zero value")
        df2.at[i, 'Property Total'] = 1
    if(burglary != 0):
        #print("Non zero value")
        df2.at[i, 'Burglary'] = 1
    if(larceny != 0):
        #print("Non zero value")
        df2.at[i, 'Larceny'] = 1
    if(motor != 0):
        #print("Non zero value")
        df2.at[i, 'Motor Vehicle Theft'] = 1

    
    
    
```

```python
df2
```

```python
df2 = apriori(df2, min_support = 0.2, use_colnames = True, verbose = 1)
df2
```

```python
df_ar = association_rules(df2, metric = "confidence", min_threshold = 0.6)
```
